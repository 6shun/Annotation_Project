# Adjucated Data
Your data with 1000 adjudicated labels.  Name this "adjudicated.txt".  This data should have four tab-separated columns:
data point ID (a unique integer)
the word "adjudicated"
the label
the original text (not tokenized)

Example:

0    adjudicated    	pos	    First off

```python
# go through rows and generate new dataframe
column_names = ["ID", "adjucated", "label", "text"]
df_adjudicated = pd.DataFrame(columns = column_names)

for r in np.arange(len(df_annotated)):
  # Get data
  id = df_annotated.loc[r,"data point ID"]
  adjudicated = 'adjudicated'
  label = df_annotated.loc[r,"adjucate_TW"]
  text = df_annotated.loc[r,"Category"] + "; " + df_annotated.loc[r,"Title"]
  # Crete new row of df
  temp = pd.DataFrame([[id, adjudicated, label, text]], columns=column_names)
  # concat on adjudicated dataframe
  df_adjudicated = pd.concat([df_adjudicated, temp])
```

```python
df_adjudicated
```

```python
df_adjudicated.to_csv('adjudicated_data.txt', sep='\t', index=False, header=False)
```

# Individual Data
Your individual annotations of the "evaluation" batch, which should include roughly 500 labels per annotator if using option a. above (Luke and Han annotate everything) or 333 labels per annotator if using option b (Luke, Han and Leia alternate being primary annotators per document).  Name this "individual_annotations.txt". This data should have four tab-separated columns:
data point ID (a unique identifier)
the annotator ID (e.g., "dbamman")
the label
the original text (not tokenized)

```python
# Only evaluation batch
df_annotated = pd.read_csv(file_path) # read original df to access evaluation batch
df_evaluation = df_annotated.iloc[500:].reset_index(drop=True)
df_evaluation

column_names = ["ID", "annotator", "label", "text"]
df_individual = pd.DataFrame(columns = column_names)


for r in np.arange(len(df_evaluation)):
  # Get data
  id = df_evaluation.loc[r,"data point ID"]

  YY = 'YangYang'
  JR = 'JingshuRui'

  label_YY = df_evaluation.loc[r,"label_YY"]
  label_JR = df_evaluation.loc[r,"label_JR"]

  text = df_evaluation.loc[r,"Category"] + "; " + df_evaluation.loc[r,"Title"]
  # Crete new row of df
  tempYY = pd.DataFrame([[id, YY, label_YY, text]], columns=column_names)
  tempJR = pd.DataFrame([[id, JR, label_JR, text]], columns=column_names)

  # concat on adjudicated dataframe
  df_individual = pd.concat([df_individual, tempYY, tempJR])
```
